```python
class Household:
    def __init__(self, data):
        self.data = data
    
    def generate_activity_chain(self):
        activity_chains = []
        
        for index, row in self.data.iterrows():
            start_time = row['STRTTIME']
            end_time = row['ENDTIME']
            duration = (end_time - start_time).seconds // 60  # Convert duration to minutes
            
            activity_chain = ''
            for i in range(6, 23, 15):  # Loop through time slots from 06:00 to 23:00
                slot_start = start_time.replace(hour=i, minute=0, second=0)
                slot_end = start_time.replace(hour=i+15, minute=0, second=0)
                
                if start_time <= slot_start < end_time or start_time < slot_end <= end_time:
                    if row['WHYFROM'] == '01' and row['WHYTO'] == '01':
                        activity_chain += 'H'  # In home
                    elif row['WHYTRP1S'] == '10':
                        activity_chain += 'W'  # Work
                    elif row['WHYTRP1S'] == '30':
                        activity_chain += 'M'  # Maintenance
                    elif row['WHYTRP1S'] == '20':
                        activity_chain += 'S'  # School
                    elif row['WHYTRP1S'] == '70':
                        activity_chain += 'K'  # Pick up/Drop off
                    elif row['WHYTRP1S'] == '50':
                        activity_chain += 'R'  # Recreational/Social
                    elif row['WHYTRP1S'] == '40':
                        activity_chain += 'P'  # Personal activities
                    else:
                        activity_chain += 'O'  # Other
            
            activity_chains.append(activity_chain)
```

```python
def convert_time(time_int):
    time_str = str(time_int)
    return time_str[:-2] + ':' + time_str[-2:]

#converitng the time to string
new_df['STRTTIME'] = new_df['STRTTIME'].apply(convert_time)
new_df['ENDTIME'] = new_df['ENDTIME'].apply(convert_time)
```
