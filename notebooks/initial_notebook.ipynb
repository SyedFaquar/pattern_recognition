```python
#converting the time into time slots for 15 minutes each from 550 to 605
import pandas as pd

# Assuming df is your DataFrame and 'STRTTIME' column contains the time in HHMM format (e.g., 550 for 5:50 AM)
# Replace 'df' with the name of your DataFrame and 'STRTTIME' with the actual column name
# Convert 'STRTTIME' to string format
new_df['STRTTIME'] = new_df['STRTTIME'].astype(str)

# Pad zeros to the left to ensure consistency (e.g., '550' becomes '0550')
new_df['STRTTIME'] = new_df['STRTTIME'].str.zfill(4)

# Convert 'STRTTIME' to datetime format
new_df['STRTTIME'] = pd.to_datetime(new_df['STRTTIME'], format='%H%M')

# Define the start and end times for the time slots
start_time = pd.to_datetime('05:50:00', format='%H:%M:%S')  # 5:50 AM
end_time = pd.to_datetime('06:05:00', format='%H:%M:%S')    # 6:05 AM

# Create a new column for time slots
new_df['TIME_SLOT'] = pd.cut(new_df['STRTTIME'], bins=pd.date_range(start_time, end_time, freq='15T'))

# Display the DataFrame with time slots
print(new_df[['STRTTIME', 'TIME_SLOT']])


#do the same for end time
# Assuming df is your DataFrame and 'STRTTIME' column contains the time in HHMM format (e.g., 550 for 5:50 AM)
# Replace 'df' with the name of your DataFrame and 'STRTTIME' with the actual column name
# Convert 'STRTTIME' to string format
new_df['ENDTIME'] = new_df['ENDTIME'].astype(str)

# Pad zeros to the left to ensure consistency (e.g., '550' becomes '0550')
new_df['ENDTIME'] = new_df['ENDTIME'].str.zfill(4)

# Convert 'STRTTIME' to datetime format
new_df['ENDTIME'] = pd.to_datetime(new_df['ENDTIME'], format='%H%M')

# Define the start and end times for the time slots
start_time = pd.to_datetime('05:50:00', format='%H:%M:%S')  # 5:50 AM
end_time = pd.to_datetime('06:05:00', format='%H:%M:%S')    # 6:05 AM

# Create a new column for time slots
new_df['END_TIME_SLOT'] = pd.cut(new_df['ENDTIME'], bins=pd.date_range(start_time, end_time, freq='15T'))

# Display the DataFrame with time slots
print(new_df[['ENDTIME', 'END_TIME_SLOT']])




```
